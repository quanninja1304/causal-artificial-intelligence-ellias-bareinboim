# what is causal AI?

## 1. Def
- Đây là khả năng nhận thức để học hỏi và khai thác các mối quan hệ nhân quả. Khác với Trí tuệ Nhân tạo hiện nay (thường chỉ giỏi tìm kiếm các khuôn mẫu tương quan trong dữ liệu), trí tuệ nhân quả tập trung vào việc hiểu "cơ chế" tạo ra dữ liệu.

- Ba hành động then chốt:

    - Học hỏi một cách hệ thống (Systematically learn): Đây là khả năng khám phá ra các thực thể trong thế giới và cách chúng liên kết với nhau ngay từ khi còn nhỏ. Thay vì chỉ ghi nhớ dữ liệu, chúng ta đặt ra các giả thuyết nhân quả và kiểm chứng chúng thông qua việc quan sát và thực hiện các can thiệp vào môi trường (giống như cách các nhà khoa học làm thí nghiệm).


    - Xử lý (Process): Khả năng mã hóa các mối quan hệ này vào trong "mô hình thế giới" nội tại của chúng ta. Điều này cho phép chúng ta suy luận để trả lời các câu hỏi về "Tại sao?" (Why?) và "Điều gì xảy ra nếu?" (What if?) .


    - Vận dụng/Khai thác (Leverage): Sử dụng các kiến thức nhân quả này để:


        + Lập kế hoạch: Chọn hành động để đạt được kết quả mong muốn.


        + Quy trách nhiệm: Xác định ai hoặc cái gì đã gây ra một sự việc (xác định lỗi hoặc công trạng).


        + Tưởng tượng: Suy nghĩ về các kịch bản không có thực (phản thực nghiệm) để rút kinh nghiệm hoặc đưa ra quyết định tốt hơn.

## **3 nấc thang Nhân quả (PCH) trả lời cho cùng một vấn đề:**
**1. Tầng 1: Quan sát (Association - "Seeing")**  
    - Câu hỏi: "Nếu tôi thấy một người uống Aspirin, khả năng họ hết đau đầu là bao nhiêu?" ($P(y \mid x)$).  
    - Hoạt động: Ở tầng này, AI chỉ đóng vai trò là một "quan sát viên" thụ động. Nó tìm kiếm các quy luật lặp đi lặp lại trong dữ liệu lịch sử.  
    - Hạn chế: Tương quan không phải là nhân quả. Ví dụ, dữ liệu có thể cho thấy người uống Aspirin thường đau đầu nặng hơn. AI Tầng 1 có thể kết luận sai lầm rằng "Aspirin gây ra đau đầu", trong khi thực tế là người ta chỉ uống Aspirin vì họ đang đau đầu. Đây là tầng mà các mô hình Deep Learning và Thống kê hiện đại đang vận hành.  

**2. Tầng 2: Can thiệp (Intervention - "Doing")**  
    - Câu hỏi: "Nếu tôi bắt bệnh nhân uống thuốc, cơn đau đầu của họ có biến mất không?" ($P(y \mid do(x))$).  
    - Hoạt động: AI lúc này trở thành một "tác nhân" chủ động. Phép toán $do(x)$ đại diện cho việc chúng ta can thiệp vào thế giới để thay đổi một biến số (ép uống thuốc) và xóa bỏ mọi yếu tố gây nhiễu bên ngoài (như việc bệnh nhân tự chọn uống thuốc vì thấy đau).  
    - Ứng dụng: Đây là nền tảng của các thử nghiệm lâm sàng ngẫu nhiên (RCT) hoặc thử nghiệm A/B trong marketing. Nó giúp xác định xem hành động của chúng ta có thực sự mang lại kết quả mong muốn hay không.  

**3. Tầng 3: Phản thực nghiệm (Counterfactuals - "Imagining")**  
    - Câu hỏi: "Bệnh nhân đã uống thuốc và vẫn bị đau đầu. Liệu nếu lúc đó họ không uống thuốc, họ có bớt đau hơn không?" ($P(y_x \mid x', y')$).  
    - Hoạt động: Đây là tầng cao nhất – khả năng "tưởng tượng". Nó không hỏi về một nhóm người (quần thể), mà hỏi về một cá nhân cụ thể trong một tình huống đã xảy ra.  
    - Tầm quan trọng: Tầng này giúp AI giải thích "tại sao" một sự việc lại xảy ra và quy trách nhiệm. Nếu không có khả năng suy luận phản thực nghiệm, AI không thể thực sự hiểu được khái niệm "lỗi" hay "nguyên nhân".  


| Layer | (Symbolic) | Typical Activity | Typical Question | Example | Machine Learning Tasks |
|------|------------|------------------|------------------|---------|------------------------|
| L1 Associational | $P(y \| x)$ | Seeing (What is?) | How would seeing X change my belief in Y? | What does a symptom tell us about the disease? | Supervised / Unsupervised Learning |
| L2 Interventional | $P(y \| do(x), c)$ | Doing (What if?) | What if I do X? | What if I take aspirin, will my headache be cured? | Reinforcement Learning |
| L3 Counterfactual | $P(y_{x} \| x', y')$ | Imagining (Why?) | What if I had acted differently? | Was it the aspirin that stopped my headache? | — |

### *Tại sao chúng ta cần đưa ***Causality*** vào cấu trúc của AI??*

**1. Nhân quả là nền tảng của Phương pháp Khoa học**  
Tác giả khẳng định rằng khoa học không chỉ là việc mô tả dữ liệu (descriptions) mà là tìm ra cơ chế (mechanisms) bên dưới.

- Quy trình khoa học: Kết hợp Quan sát + Thí nghiệm -> Đưa ra giả thuyết về cơ chế nhân quả chưa biết -> Kiểm chứng lại bằng thực nghiệm.

- Ví dụ Sinh học: Các nhà khoa học giải thích việc tổng hợp protein bằng một chuỗi các bước nhân quả từ DNA -> mRNA -> Protein.

- Ví dụ Kinh tế: Dự báo hành vi tiêu dùng dựa trên các biến số như thu nhập, tài sản và lãi suất.

- Điểm cốt lõi: Khi hiểu được cơ chế, chúng ta có thể dự đoán điều gì sẽ xảy ra trong các kịch bản giả định (hypothetical) mà chúng ta chưa từng thấy trong quá khứ (ví dụ: tác động của một loại virus mới hoặc một chính sách lãi suất chưa từng có).

**2. Sự "Toán học hóa" Nhân quả (Sự tương đồng với Alan Turing)**  
Đây là một quan điểm lịch sử rất thú vị:

- Trước những năm 1920: Khái niệm "tính toán" (computation) chỉ được hiểu một cách mơ hồ qua trực giác. Cho đến khi Alan Turing và các đồng nghiệp đưa ra các mô hình toán học và thuật toán chính xác, ngành Khoa học Máy tính mới thực sự bùng nổ và tạo ra thế giới công nghệ như ngày nay.

- Hiện nay: Khái niệm "nhân quả" cũng đang trải qua một quá trình tương tự. Trước đây nó chỉ là triết học và trực giác, nhưng vài thập kỷ qua, nó đã được toán học hóa (thông qua ngôn ngữ lập trình, thuật toán và phân tích độ phức tạp).

- Thông điệp: Chúng ta đang ở thời điểm "Turing" của nhân quả. Việc toán học hóa này sẽ dẫn đến một thế hệ kỹ thuật mới: Causal AI.

**3. Định nghĩa về Causal AI (Trí tuệ nhân tạo nhân quả)**  
Tác giả định nghĩa Causal AI không chỉ là các hệ thống biết dự đoán (prediction) mà phải có khả năng:

- Nhận diện các mối quan hệ nhân quả: Biết cái gì gây ra cái gì.

- Hành động dựa trên các lựa chọn phản thực nghiệm: Biết suy tính "nếu mình làm khác đi thì sao".

- Bền vững (Robust): Vẫn hoạt động chính xác ngay cả khi các điều kiện môi trường thay đổi (điều mà AI dựa trên tương quan thường thất bại).

**4. Các mục tiêu cốt lõi (Causal-AI Desiderata)**  
Cuốn sách hướng tới việc xây dựng các hệ thống AI đạt được 5 tiêu chuẩn:

- Safety (An toàn): AI hiểu được hậu quả của hành động.

- Robustness (Bền vững): Không dễ bị sai lệch khi dữ liệu đầu vào thay đổi.

- Human Compatibility (Tương thích với con người): Có cách tư duy giống con người.

- Explainability (Có thể giải thích): Giải thích được tại sao nó đưa ra quyết định đó.

- Social Alignment (Phù hợp với xã hội): Đảm bảo tính công bằng và đạo đức.

### 1.3 the causal approach to ai

**1.3.1. modeling the agent-environment relationship through causal language.**  
-> Làm thế nào để một thực thể (Agent) hiểu và vận hành trong một thế giới phức tạp (Environment)?  
**1.3.2. 5 thách thức của AI (Grand Challenges)**  

- Thách thức 1: Hiểu và Giải thích (Reasoning & Explanations)
    - Vấn đề: Để giải thích một sự việc, ta phải biết lịch sử nhân quả của nó.

    - Ví dụ: Bác sĩ không chỉ cần biết bệnh nhân có triệu chứng, mà phải biết "tại sao" triệu chứng đó xuất hiện (do vi khuẩn, chấn thương hay gen?). Giải thích chính là cung cấp thông tin về lịch sử nhân quả.

- Thách thức 2: Ra quyết định chính xác và hiệu quả (Decision-making)
    - Vấn đề: Các hệ thống Học tăng cường (Reinforcement Learning - RL) hiện nay cần hàng triệu lần thử sai mới học được một nhiệm vụ đơn giản vì chúng không hiểu cấu trúc nhân quả của môi trường.

    - Rào cản đạo đức: Trong y tế, ta không thể "thử sai" trên người (ví dụ: bắt người ta hút thuốc để xem có bị ung thư không). Causal AI giúp AI đưa ra quyết định tốt mà không cần thử nghiệm mù quáng quá nhiều.

- Thách thức 3: Tính Tổng quát hóa và Bền vững (Generalizability & Robustness)
    - Vấn đề: AI thường học rất giỏi trên dữ liệu huấn luyện nhưng thất bại khi môi trường thực tế thay đổi một chút.

    - Giải pháp: Các cơ chế nhân quả thường mang tính bất biến (invariance). Nếu AI học được cơ chế nhân quả thay vì các tương quan bề mặt, nó sẽ hoạt động ổn định trong nhiều môi trường khác nhau.

- Thách thức 4: Tạo dựng thế giới phản thực nghiệm (Counterfactual Generation)
    - Vấn đề: Các mô hình tạo sinh (như LLM hay AI vẽ tranh) hiện nay chủ yếu là "bắt chước" (mimicking) dữ liệu quan sát.

    - Giải pháp: Một mô hình tạo sinh thực sự phải có khả năng mô phỏng các thực tại khác nhau qua 3 tầng của PCH, giúp ta trả lời câu hỏi "Nếu điều này thay đổi thì kết quả sẽ trông như thế nào?".

- Thách thức 5: Học cấu trúc và biến số nhân quả (Model Learning & Discovery)
    - Vấn đề: Làm sao để AI tự khám phá ra các quy luật nhân quả từ dữ liệu thô (hình ảnh, video, văn bản) thay vì phải có con người nạp sẵn biểu đồ nhân quả?

    - Giải pháp: Đây là tầng sâu nhất, giúp AI tự xây dựng "World Model" (Mô hình thế giới) của riêng nó từ những quan sát thực tế.  

**1.3.3. Current AI through Causal LensesThe structural**  
- AI truyền thống: Hoạt động bên trong một "vòng lặp lấy mẫu" (Data và Query cùng nằm ở một tầng).

- Causal AI: Cho phép chúng ta thực hiện các suy luận xuyên tầng. Ví dụ: Dùng dữ liệu quan sát (L1) để trả lời các câu hỏi về can thiệp (L2) hoặc phản thực nghiệm (L3).
