# **Chapter 3: A Structural Account for Causal Understanding**

## **3.1 Understanding in Science**

### 1. Mục tiêu của khoa học: Hiểu (Understanding)
- Một mục tiêu cốt lõi của khoa học là **mở “hộp đen” của tự nhiên**, giải thích:
  - Thế giới vận hành như thế nào
  - Vì sao các hiện tượng xảy ra theo cách chúng xảy ra
- **Giải thích khoa học** không chỉ mô tả hiện tượng phức tạp, mà phải:
  - Quy giản (reduce) chúng về các **khối xây dựng cơ bản** (elementary building blocks)
  - Có thể hiểu được từ các **nguyên lý nền tảng (first principles)**

> Tinh thần của Einstein: *Nếu bạn không thể giải thích nó một cách đơn giản, bạn chưa thực sự hiểu nó.*

---

### 2. Điều tra khoa học như quá trình tìm kiếm cơ chế nhân quả
- Khoa học tiến bộ thông qua:
  - **Quan sát + thực nghiệm**
  - **Lập luận logic**
- Mục tiêu sâu xa là **xác định các cơ chế nhân quả (causal mechanisms)**

#### Ví dụ điển hình
- **Sinh học phân tử**:
  - Giải thích quá trình tổng hợp protein thông qua DNA, mRNA, tRNA, codon, amino acid
  - Có thể dự đoán tác động của virus/vi khuẩn lên các pathway sinh học
- **Vật lý hiện đại**:
  - Mô hình Chuẩn (Standard Model): fermion, boson, các lực cơ bản
  - Thuyết tương đối rộng: sao, hố đen, sóng hấp dẫn

-> Điểm chung:  
Xây dựng **mô tả cô đọng (parsimonious)** của hiện tượng tự nhiên bằng các thành phần cơ bản, có thể:
- Hiểu
- Phản biện
- Ứng dụng để phát triển công nghệ

---

### 3. Đặc điểm của quá trình khám phá khoa học
Một mô hình/ký thuyết khoa học tốt cần thỏa mãn:

#### (a) Mô hình hóa (Modeling)
- Xuất phát từ:
  - Sự sáng tạo
  - Hiểu biết sâu về bối cảnh và thực tại không quan sát được phía sau dữ liệu

#### (b) Giải thích dữ liệu đã quan sát
- Phải **phù hợp với dữ liệu trong quá khứ**
- Tương ứng với:
  - Layer 1 (observational)
  - Layer 2 (interventional) trong **Causal Hierarchy (PCH)**

#### (c) Dự đoán mới
- Tạo ra **dự đoán chưa từng xuất hiện trong dữ liệu**
- Đây là yêu cầu khó nhất và quan trọng nhất
- Thể hiện rằng mô hình đã nắm bắt được **cấu trúc nền tảng của hiện tượng**

-> (b) thường đạt được nếu mô hình đủ phức tạp  
-> (c) mới là **chuẩn vàng (gold standard)** của hiểu biết khoa học lâu dài

---

### 4. Mô tả chuẩn của khám phá khoa học bằng SCM
- Hiện tượng mục tiêu cần giải thích
- Ta xây dựng một mô hình đầy đủ gọi là **$M^*$**
  - $M^*$ ∈ $\Omega$ (không gian tất cả các mô hình khả dĩ)
  - Thường là một **Structural Causal Model (SCM)** hoàn chỉnh
  - Bao gồm:
    - Các hàm cấu trúc $F$
    - Phân phối nhiễu ngoại sinh $P(U)$

#### Hai yêu cầu đối với $M^*$
1. **Phù hợp với dữ liệu $D$** (observations + experiments)
2. **Cho phép suy ra dự đoán mới** (interventional, counterfactual)

-> Khi có $M^*$, ta có thể:
- Mô phỏng mọi truy vấn:
  - Quan sát
  - Can thiệp
  - Phản thực (counterfactual)

---

### 5. Hạn chế trong AI, khoa học xã hội và khoa học dữ liệu
- Không thể đạt mức hiểu biết như vật lý hay sinh học cơ bản do:
  - Con người nằm trong vòng lặp
  - Hệ thống mở, phức tạp
  - Nhiễu và ngẫu nhiên không thể mô hình hóa triệt để
- Quan sát này dẫn đến:
  - Việc **áp dụng xác suất** rộng rãi trong khoa học

#### Vấn đề cốt lõi
- Dữ liệu quan sát $P(V)$ (Layer 1) **không đủ** để:
  - Suy ra tác động can thiệp (Layer 2)
  - Suy luận phản thực (Layer 3)
- Đây là nội dung của **Causal Hierarchy Theorem (CHT)**

-> Phân phối xác suất là **quá thô** để nắm bắt cấu trúc nhân quả

---

### 6. Cách tiếp cận “hiểu biết” thay thế trong nhân quả
- Thừa nhận:
  - $M^*$ tồn tại và sinh dữ liệu
  - Nhưng **không thể học trực tiếp**
- Giải pháp:
  - Áp đặt **ràng buộc định tính (qualitative constraints)** lên không gian mô hình $\Omega$

#### Vai trò của sơ đồ nhân quả $G$
- $G$ là:
  - Một **mã hóa cô đọng** của hiểu biết nhân quả
  - Xác định:
    - Thứ tự nhân quả giữa các biến nội sinh
    - Quan hệ độc lập giữa các nhiễu ngoại sinh
- $G$ xác định tập con $\Omega_G \subset \Omega$
- Không cam kết:
  - Dạng hàm cụ thể
  - Phân phối cụ thể của $U$

---

### 7. Kiểm chứng và suy luận nhân quả
- **Yêu cầu 1 – Phù hợp dữ liệu**:
  - Kiểm tra bằng **d-separation**
  - Falsifiability: mô hình có thể bị bác bỏ bởi dữ liệu
- **Yêu cầu 2 – Dự đoán mới**:
  - Suy luận L2, L3 từ:
    - Dữ liệu
    - Giả định được mã hóa trong $G$

-> **Suy luận nhân quả (causal reasoning)**:
- Kết hợp:
  1. Dữ liệu quá khứ
  2. Mô hình (tập giả định)
- Để tạo ra:
  - Hiểu biết mới về thế giới
  - Các dự đoán ngoài dữ liệu quan sát

---

### 8. Tổng kết
- Hiểu biết khoa học ≠ chỉ khớp dữ liệu
- Hiểu biết khoa học =  
  **Mô hình hóa cấu trúc nhân quả + khả năng dự đoán những kịch bản chưa quan sát**
- Trong AI và khoa học dữ liệu:
  - Sơ đồ nhân quả $G$ đóng vai trò trung tâm
  - Là cầu nối giữa dữ liệu và suy luận về các “thế giới thay thế”


## 3.2 Causal Understanding through Causal Reasoning.

### 3.2.1 Hình thức hóa Sự Thấu hiểu Nhân quả (Formalizing Causal Understanding)

Trong phần này, chúng ta xây dựng một **bộ máy hình thức** nhằm làm rõ cách các đại lượng trong **Pearl’s Causal Hierarchy (PCH)** được so sánh với nhau về **Causal Informativeness** (mức độ giàu thông tin nhân quả).

---

#### 1. Quy trình sinh dữ liệu hai bước (Two-step Generative Process)

Mọi biến thiên quan sát được của các **biến nội sinh** (Endogenous variables) $V$ đều xuất phát từ một quy trình sinh dữ liệu gồm hai bước:

**Bước 1 – Lấy mẫu (Sampling step)**  
Một đơn vị (unit) $U = u$ được rút ra từ quần thể theo phân phối xác suất $P(U)$.  
$U$ đại diện cho các **yếu tố ngoại sinh** (Exogenous factors) không quan sát được, ví dụ: gen di truyền, điều kiện kinh tế – xã hội, đặc điểm bẩm sinh.

**Bước 2 – Đánh giá (Evaluation step)**  
Đơn vị $u$ sau đó đi qua một chuỗi các **cơ chế cấu trúc** (Structural mechanisms) $F_{\text{exec}}$ theo thứ tự nhân quả để xác định giá trị của $V$.

- Nếu $F_{\text{exec}} = F$ (cơ chế tự nhiên) → thu được **Observational Distribution** (phân phối quan sát).
- Nếu $F_{\text{exec}} = F_{X=x}$ (can thiệp) → thu được **Interventional Regime** (chế độ can thiệp).

---

#### 2. Định nghĩa: Contrast Thực tế – Phản thực tế (Definition 3.2.1)

Để so sánh các kết quả, ta sử dụng khái niệm **Contrast (sự tương phản)**:

$$
C(C_0, C_1, E_0, E_1)
= P(y_{C_1} \mid E_1) - P(y_{C_0} \mid E_0)
$$

Trong đó:

- $E_0, E_1$ là **Factual clauses** (mệnh đề thực tế): những gì đã được quan sát.
- $C_0, C_1$ là **Counterfactual clauses** (mệnh đề phản thực tế): các can thiệp mà biến kết quả $Y$ phản ứng lại.

Các trường hợp đặc biệt:
- Nếu $E_0 = E_1$: Contrast hoàn toàn là **Counterfactual** (chỉ thay đổi can thiệp trên cùng một nhóm đối tượng).
- Nếu $C_0 = C_1$: Contrast hoàn toàn là **Factual** (chỉ thay đổi nhóm đối tượng dưới cùng một cơ chế).

---

#### 3. Khai triển Cơ sở Cấu trúc (Structural Basis Expansion)

> **Theorem 3.2.1 — Contrast’s Decomposition & Structural Basis Expansion.**
>
> Let $M$ be an SCM and $C(C_0, C_1, E_0, E_1)$ be a contrast $P(y_{C_1} | E_1) - P(y_{C_0} | E_0)$. $C$ can be decomposed into its counterfactual and factual variations, namely:
>
> $$C = \underbrace{P(y_{C_1} | E_1) - P(y_{C_0} | E_1)}_{\text{counterfactual contrast}} + \underbrace{P(y_{C_0} | E_1) - P(y_{C_0} | E_0)}_{\text{factual contrast}} \quad (3.2)$$
>
> Furthermore, the corresponding counterfactual and factual contrasts admit the following structural basis expansions in terms of $M$, respectively:
>
> (a) **Counterfactual contrast ($C_{ctf}$)**, where $E_0 = E_1 = E$, can be expanded as:
> $$P(y_{C_1} | E) - P(y_{C_0} | E) = \sum_{u} \underbrace{(y_{C_1}(u) - y_{C_0}(u))}_{\text{unit-level difference}} \cdot \underbrace{P(u | E)}_{\text{posterior}} \quad (3.3)$$
>
> (b) **Factual contrast ($C_{factual}$)**, where $C_0 = C_1 = C$, can be expanded as:
> $$P(y_C | E_1) - P(y_C | E_0) = \sum_{u} \underbrace{y_C(u)}_{\text{unit outcome}} \cdot \underbrace{(P(u | E_1) - P(u | E_0))}_{\text{posterior difference}} \quad (3.4)$$

##### A. Counterfactual Contrast ($C_{\text{ctf}}$)

Đo lường sự khác biệt về phản ứng giữa hai can thiệp $C_1$ và $C_0$ trên **cùng một tập bằng chứng** $E$:

$$
P(y_{C_1} \mid E) - P(y_{C_0} \mid E)
= \sum_u [y_{C_1}(u) - y_{C_0}(u)] \times P(u \mid E)
$$

- $[y_{C_1}(u) - y_{C_0}(u)]$: **Individual Causal Effect (ICE)*** khác biệt ở **cấp độ đơn vị (unit-level)**, do cơ chế cấu trúc $F$ quyết định.
- $P(u \mid E)$: **xác suất hậu nghiệm (posterior)**, xác định tập con quần thể (subpopulation) dựa trên bằng chứng $E$.
> Counterfactual Contrast là kì vọng có điều kiện theo phân phối hậu nghiệm (posterior) của $U$ của hiệu ứng nhân quả cá nhân (ICE).  
> $$
> C_{ctf} (C_1, C_0, E) = \mathbf{E}_{U \mid E} [Y_{C_1}, Y_{C_0}]
> $$
> $\quad$

##### B. Factual Contrast ($C_{\text{factual}}$)

Đo lường sự khác biệt giữa hai nhóm đối tượng $E_1$ và $E_0$ dưới **cùng một can thiệp** $C$:

$$
P(y_C \mid E_1) - P(y_C \mid E_0)
= \sum_u y_C(u) \times [P(u \mid E_1) - P(u \mid E_0)]
$$

Sự biến thiên ở đây đến từ **khác biệt trong phân phối quần thể** (posterior difference).  
Đây chính là nguồn gốc của các **tương quan giả** (spurious associations).

---

### 3.2.2 Các cấp độ Thấu hiểu Nhân quả  
#### (CUP – Causal-Understanding Plane)

Dựa trên khai triển cơ sở cấu trúc, ta xây dựng **Mặt phẳng Thấu hiểu Nhân quả (CUP)** với hai trục chính:

- **Trục Cơ chế (Mechanisms – trục X)**:  
  Từ giữ nguyên cơ chế tự nhiên (observational) đến thay đổi cơ chế (interventional).
- **Trục Quần thể (Population – trục Y)**:  
  Từ lấy mẫu ngẫu nhiên toàn quần thể đến chọn lọc nhóm con (subpopulations) hoặc từng cá nhân (unit-level).

#### So sánh các tầng của PCH qua CUP

| Layer | Đại lượng tiêu biểu | Mệnh đề (Contrast) | Quyền năng giải thích |
|------|--------------------|--------------------|----------------------|
| L1 | TV (Total Variation) | $\langle \emptyset, \emptyset, x_0, x_1 \rangle$ | Thấp. Trộn lẫn giữa cơ chế tự nhiên và khác biệt quần thể. |
| L2 | TE (Total Effect) | $\langle x_0, x_1, \emptyset, \emptyset \rangle$ | Trung bình. Cô lập tác động của cơ chế trên trung bình quần thể. |
| L2z | Conditional TE | $\langle x_0, x_1, z, z \rangle$ | Cao hơn TE. Tác động can thiệp trên nhóm cụ thể $Z=z$. |
| L3 | ETT (Effect of Treatment on the Treated) | $\langle x_0, x_1, x, x \rangle$ | Rất cao. Phân tích nhóm có xu hướng tự nhiên chọn $X=x$. |
| L3 | PN / PS (Necessity / Sufficiency) | $\langle x_0, x_1, \{x,y\}, \{x,y\} \rangle$ | Cực cao. Kết hợp can thiệp và bằng chứng về kết quả thực tế $Y=y$. |

---

#### Tại sao L3 mạnh hơn L2?

Khác biệt cốt lõi giữa L2 và L3 nằm ở **sự xung đột (conflict)**.  
L3 xem xét các đơn vị có xu hướng tự nhiên chọn $X = x_1$ nhưng lại bị **ép buộc (can thiệp)** vào điều kiện $X = x_0$.

Khả năng xử lý mâu thuẫn giữa:
- *“cái đã xảy ra”* và  
- *“cái lẽ ra nên xảy ra”*  

là chìa khóa để hiểu sâu hơn về **trách nhiệm (blame)** và **giải thích nhân quả (causal explanation)**.

Ví dụ

### 1. Sinh mô hình nhân quả (SCM)
Giả sử:
* $U$: Khả năng tiềm ẩn (latent ability) - Biến ngoại sinh.
* $X$: Quyết định điều trị (tự chọn $\rightarrow$ confounding).
* $Y$: Kết quả đầu ra.

```python
import numpy as np

np.random.seed(42)
N = 100_000

# Exogenous variable (population heterogeneity)
U = np.random.normal(0, 1, N)

# Treatment assignment (confounded)
# People with higher U are more likely to choose X=1
p_X = 1 / (1 + np.exp(-U))
X = np.random.binomial(1, p_X)

# Structural equation for Y
def f_Y(x, u):
    noise = np.random.normal(0, 0.3, len(u))
    return (1.5 * x + u + noise > 0).astype(int)

# Observed outcome
Y = f_Y(X, U)

# Counterfactual outcomes
Y_x1 = f_Y(np.ones(N), U)
Y_x0 = f_Y(np.zeros(N), U)
```

#### 2. L1 - Total Variation (Observational)
$$
    TV =P(Y=1 \mid X=1) - P(Y=1 \mid X=0)
$$

```python
TV = Y[X == 1].mean() - Y[X == 0].mean()
print("L1 — Total Variation:", TV)
```

#### **L2 — Total Effect (ATE)**
$$
TE = \mathbf{E} [Y_{x_1}] - \mathbf{E} [Y_{x_0}]
$$

```python
TE = Y_x1.mean() - Y_x0.mean()
print("L2 — Total Effect (ATE):", TE)
```

#### **L2z — Conditional TE (CATE)**
$$
CATE = \mathbf{E} [Y_{x_1} - Y_{x_0} \mid U > 0]
$$ 

```python
mask = U > 0

CATE = Y_x1[mask].mean() - Y_x0[mask].mean()
print("L2z — CATE (U > 0):", CATE)
```

#### **L3 — ETT (Effect of Treatment on the Treated)**
$$
ETT = \mathbf{E} [Y_{x_1} - Y_{x_0} \mid X=1]

```python
# NOTe: Quần thể được định nghĩa bởi hành vi tự nhiên ($X=1$). Xuất hiện xung đột giữa thực tế ($factual: X=1$) và giả định ($counterfactual: Y_{x_0}$).
ETT = (Y_x1[X == 1] - Y_x0[X == 1]).mean()
print("L3 — ETT:", ETT)
```

#### **L3 — PN (Probability of Necessity)**
$$
PN = \mathbf{P} (Y_{x_0} = 0 \mid X=1, Y=1)
$$


```python
mask = (X == 1) & (Y == 1)
PN = (Y_x0[mask] == 0).mean()
print("L3 — Probability of Necessity (PN):", PN)
```

```python
print("\nCUP Summary")
print(f"L1  TV   : {TV:.3f}")
print(f"L2  TE   : {TE:.3f}")
print(f"L2z CATE : {CATE:.3f}")
print(f"L3  ETT  : {ETT:.3f}")
print(f"L3  PN   : {PN:.3f}")
```

| Layer | Điều gì thay đổi | Thấy được gì |
| :--- | :--- | :--- |
| **L1** | Không can thiệp, không kiểm soát | Sự tương quan (Correlation) |
| **L2** | Can thiệp (Intervention) | Cơ chế nhân quả trung bình |
| **L2z** | Can thiệp + nhóm con (Sub-group) | Sự dị biệt nhân quả (Heterogeneity) |
| **L3** | Can thiệp + nhóm nội sinh (Endogenous selection) | Trách nhiệm/Đóng góp (Responsibility) |
| **L3 (PN)** | Can thiệp + Nhóm nội sinh + Kết quả thực tế | Giải thích mức độ cá thể |
